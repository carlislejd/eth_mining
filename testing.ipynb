```python
import os

from os import getenv
from dotenv import load_dotenv
from web3 import Web3, HTTPProvider

load_dotenv()
```

```python
w3 = Web3(HTTPProvider(f"https://mainnet.infura.io/v3/{getenv('INFURA_KEY')}"))
```

```python
end_block = w3.eth.blockNumber
start_block = end_block - 5
```

```python
txs = []
for x in range(start_block, end_block):
    block = w3.eth.get_block(x, True)
    for tx in block.transactions:
        txs.append(tx)
```

```python
w3.eth.get_transaction(txs[0].hash)
```
